# Прогнозирование для LPM-модели регрессии

```python
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols
```

## labour force equation

Для датасета [mroz_Greene](http://www.stern.nyu.edu/~wgreene/Text/Edition7/TableF5-1.csv) рассморим регрессию **LFP на WA, WA^2, CIT, UN, log(FAMINC)**

```python
# подключим датасет mroz_Greene 
df = pd.read_csv('../../datasets/TableF5-1.csv')
```

```python
# Создаем спецификацию модели через формулу и подгоняем модель с поправкой на гетероскедастичность
res_hc = ols(formula='LFP ~ 1 + WA + I(WA ** 2) + CIT + UN + np.log(FAMINC)', data=df).fit(cov_type='HC3')
print(res_hc.summary(slim=True))
```

Рассморим трёх людей с характеристиками

```python
new_df = pd.DataFrame( {'WA': [34, 36, 42], 
                      'CIT': [1, 0, 0], 
                     'UN': [3, 5, 7.5], 
                     'FAMINC': [35000, 48500, 67800]} )
new_df 
```

### Построим прогноз для каждого человека по подогнанной модели

```python
res_hc.predict(exog=new_df, transform=True).round(3)
```
