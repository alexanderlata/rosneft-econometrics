# logit/probit/LPM-регрессия: прогнозирование

```python
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols, logit, probit
from statsmodels.iolib.summary2 import summary_col

# Не показывать FutureWarnings
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

## labour force equation

Для датасета [mroz_Green](http://www.stern.nyu.edu/~wgreene/Text/Edition7/TableF5-1.csv)
рассморим регрессию **LFP на WA, log(FAMINC), WE, KL6, K618, CIT, UN**
трёх спецификаций:

- LPM
- logit
- probit

```python
# подключим датасет mroz_Greene
df = pd.read_csv('TableF5-1.csv')
```

## Спецификация и подгонка

```python
# LPM 
res_lpm_hc = ols(formula='LFP~WA+np.log(FAMINC)+WE+KL6+K618+CIT+UN', data=df).fit(cov_type='HC3')
```

```python
# logit
res_logit = logit(formula='LFP~WA+np.log(FAMINC)+WE+KL6+K618+CIT+UN', data=df).fit()
```

```python
# probit
res_probit = probit(formula='LFP~WA+np.log(FAMINC)+WE+KL6+K618+CIT+UN', data=df).fit()
```

```python
# Сравнение моделей
print(summary_col([res_lpm_hc, res_logit, res_probit], 
                  model_names=['Robust LPM', 'logit', 'probit'], 
                  float_format="%.3f", 
                  stars=True, 
                  info_dict = {'Observations': lambda x: '{0:d}'.format(int(x.nobs)), 
                               'Log Likelihood': lambda x: '{:.3f}'.format(x.llf), 
                               'Akaike Inf. Crit.': lambda x: '{:.3f}'.format(x.aic)
                               }))
```

## Прогнозирование

```python
# Создадим датафрейм с новыми данными регрессоров для прогноза
new_data = pd.DataFrame({'WA':[35, 40, 42], 
                         'FAMINC':[12500, 9800, 67800], 
                         'WE': [15, 12, 14], 
                         'KL6': [2, 1, 2],
                         'K618': [0, 2, 1], 
                         'CIT': [1, 0, 1], 
                         'UN':[5, 7.5, 3]})
new_data
```

```python
# Прогноз для LPM с округлением до 3-х десятичных знаков
res_lpm_hc.predict(exog=new_data, transform=True).round(3)
```

```python
# Прогноз для logit с округлением до 3-х десятичных знаков
res_logit.predict(exog=new_data, transform=True).round(3)
```

```python
# Прогноз для probit с округлением до 3-х десятичных знаков
res_probit.predict(exog=new_data, transform=True).round(3)
```
